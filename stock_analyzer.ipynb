```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning) 
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt 

# Specify the stock ticker
ticker = "WFC"

# Fetch stock data (adjust the period as needed)
data = yf.download(ticker, start="2023-02-01", end="2024-02-16")

# Focus on closing price
closing_prices = data['Close']

# Visualize the closing price history
plt.figure(figsize=(10, 5)) 
plt.plot(closing_prices)
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.title(f"Closing Price History for {ticker}")
plt.show() 

# Calculate Simple Moving Averages (SMA)
sma_20 = closing_prices.rolling(window=20).mean()  # 20-day SMA
sma_50 = closing_prices.rolling(window=50).mean()  # 50-day SMA

# Plot closing price along with SMAs
plt.figure(figsize=(10, 5))  
plt.plot(closing_prices, label="Closing Price")
plt.plot(sma_20, label="20-day SMA")
plt.plot(sma_50, label="50-day SMA")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title(f"Price and Moving Averages for {ticker}")
plt.legend()  # Show the legend
plt.show()

# Calculate and print basic statistics
print("Basic Statistics:")
print("Mean:", closing_prices.mean())
print("Standard Deviation:", closing_prices.std())
```
